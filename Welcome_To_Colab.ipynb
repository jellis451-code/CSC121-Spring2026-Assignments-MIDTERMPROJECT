```
# NAME: rentalcar.py

def validate_classification(code):
    while code.upper() not in ('B', 'S', 'L'):
        print("Invalid classification. Please enter B, S, or L.")
        code = input("Enter classification code (B, S, L): ")
    return code.upper()


def get_classification_name(code):
    if code == 'B':
        return "Budget"
    elif code == 'S':
        return "Standard"
    else:
        return "Luxury"


def calculate_miles(start, end):
    return end - start


def calculate_charged_miles(total_miles):
    if total_miles > 100:
        return total_miles - 100
    return 0


# FRUITFUL FUNCTION (NO PRINTS)
def calculate_bill(code, days, charged_miles):
    daily_rates = {
        'B': 40,
        'S': 60,
        'L': 80
    }

    daily_cost = daily_rates[code] * days
    mileage_cost = charged_miles * 0.35
    return daily_cost + mileage_cost


# PROGRAM LOGIC
code = input("Enter classification code (B, S, L): ")
code = validate_classification(code)

start_odometer = int(input("Start Odometer: "))
end_odometer = int(input("End Odometer: "))
days = int(input("Days Rented: "))

total_miles = calculate_miles(start_odometer, end_odometer)
charged_miles = calculate_charged_miles(total_miles)
total_bill = calculate_bill(code, days, charged_miles)
classification_name = get_classification_name(code)

print("\n--- RENTAL SUMMARY ---")
print(f"Classification: {classification_name}")
print(f"Total Miles: {total_miles}")
print(f"Charged Miles: {charged_miles}")
print(f"Total Bill: ${total_bill:.2f}")
```
